# **Filtering a DataFrame**

**Some Question on the filternig data frame
1. find  all the final winner**

```python
ipl.head(2)
```

```python
#find all the final winner
match=ipl['MatchNumber']=='Final'
new_df=ipl[match]
new_df[['Season','WinningTeam']].value_counts()
```

```python
from pandas._libs.hashtable import value_count
ipl[ipl['MatchNumber']=='Final'][["Season","WinningTeam"]]
```

```python
# how many super over finishes have occured
ipl[ipl['SuperOver'] == "Y"].shape[0]
```

```python
# how many matches has csk won in kolkata
ipl[(ipl["City"]== "Kolkata") & (ipl['WinningTeam']=='Chennai Super Kings')].shape[0]
```

```python
#toss winner is match winner in percentage
ipl[ipl["TossWinner"] == ipl['WinningTeam']].shape[0]/ipl.shape[0]*100
```

```python
# movie with raing than 8 and votes > 10000
movie[(movie["imdb_rating"] > 8) & (movie["imdb_votes"] > 10000)].shape[0]
```

**Action movie with rating higher than 7.5**

```python
#Action movie with rating higher than 7.5
mov1 = movie['genres'].str.split('|').apply(lambda x: 'Action' in x)
mov2 = movie['imdb_rating'] > 7.5
movie[mov1 & mov2].shape
```

```python
mov1 = movie['genres'].str.contains('Action')
mov2 = movie['imdb_rating'] > 7.5
movie[mov1 & mov2].shape
```

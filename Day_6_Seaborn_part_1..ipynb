# **Import Necessary library**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

# **Relational Plot**

```python
tips=sns.load_dataset('tips')
tips.head()
```

**Seaborn Scatter plot**

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="sex")
```

**We can also create the scatter plot using figure level function with the help of "relplot"**

```python
sns.relplot(data=tips,x="total_bill",y="tip",kind="scatter")
```

**use the hue, style and size parameter for better visual**

```python
sns.relplot(data=tips,x="total_bill",y="tip",kind="scatter",hue="sex",style="time",size="size")
```

```python
#using relplot to create scatter plot 
sns.relplot(data=tips,x="total_bill",y="tip",hue="sex",style="time",size="size")
```

# **Line plot**

```python
gap = px.data.gapminder()
gap.head()
```

```python
new_gap=gap[gap["country"]=="India"]
new_gap.head()
```

```python
#  create this plot using axes level function 
sns.lineplot(data=new_gap,x="year",y="lifeExp")
```

```python
#Create this plot using figure level function with the help of "replot"
sns.relplot(data=new_gap,x="year",y="lifeExp",kind="line")
```

```python
gap_new=gap[gap["country"].isin(["India","Pakistan","China"])]
gap_new.head()
```

```python
sns.relplot(kind="line",data=gap_new,x="year",y="lifeExp",hue="country")
```

```python
sns.lineplot(data=gap_new,x="year",y="lifeExp",hue="country")
```

```python
gap_new1=gap[gap["country"].isin(["India","Brazil","Germany"])]
gap_new1.head()
```

```python
sns.relplot(kind="line",data=gap_new1,x="year",y="lifeExp",hue="country",style="continent",size="continent")
```

# **Heatmap**

```python
df = gap.pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(15,13))
sns.heatmap(df)
```

```python
new_gap=gap[gap["continent"]=="Europe"]
new_gap.head()
```

```python
df = new_gap.pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(15,13))
sns.heatmap(df)
```

**use annot parameter for show the each element of the heat map**

```python
df = new_gap.pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(15,13))
sns.heatmap(df,annot=True)
```

```python
plt.figure(figsize=(15,9))
sns.heatmap(df,annot=True)
```

**linewidths parameter to show better heatmap**

```python
plt.figure(figsize=(15,15))
sns.heatmap(df,annot=True,linewidths=0.5)
```

**change the color**

```python
plt.figure(figsize=(15,15))
sns.heatmap(df,annot=True,linewidths=0.5,cmap="jet")
```

**use the "sharex" parameter for overploting text**

```python
fig,ax =plt.subplots(nrows=2,ncols=1,figsize=(10,8),sharex=True)
ax[0].scatter(better["avg"],better["strike_rate"],color="Red")
ax[1].scatter(better["avg"],better["runs"],color="green")

ax[0].set_title("This is avg V/S strike_rate")
ax[0].set_xlabel("Avg")
ax[0].set_ylabel("Strike_rate")

ax[1].set_title("This is avg V/S runs")
ax[1].set_xlabel("Avg")
ax[1].set_ylabel("Run's")
```
